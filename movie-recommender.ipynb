```python
import numpy as np
import pandas as pd
import ast
```

```python
# storing data in variables
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
# checking data parameters
movies.head(1)
```

```python
credits.head(1)
```

```python
# merging datasets on the basis of title
movies = movies.merge(credits, on = "title")
```

```python
movies.head(1)
```

```python
# filtering columns for relevance
# keeping: genres, id, keywords, title, overview, cast, crew
movies = movies[['movie_id', 'genres', 'overview', 'title', 'keywords', 'cast', 'crew']]
```

```python
movies.head()
```

```python
# merging several columns for "tags"
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
# dropped missing data
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# need to change format to a list of strings, rather than a list of dictionaries
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
        
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
# keywords, genres are now clean
```

```python
movies['cast'][0]
```

```python
# I will only consider top 3 actors in the cast
def top3(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count != 3: 
        	L.append(i['name'])
        	count += 1
        else: 
            break
    return L
```

```python
movies['cast'].apply(top3)
```

```python
movies['cast'] = movies['cast'].apply(top3)
movies.head()
```

```python
# cast is also clean now
movies['crew'][0]
```

```python
# only director matters in crew, hence I will try to extract the designation of 'director' only
def fetch_director(obj): 
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L

        
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
# crew clearly consists of the director now
movies['overview'][0]
```

```python
# converting string of overview to a list of strings
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# removing spaces from names to avoid ambiguity
# Zoe Saldana -> ZoeSaldana
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies['cast'][0]
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
movies.head()
```

```python
# data is all clean now, merging in tags now
movies['tags'] = movies['genres']+movies['keywords']+movies['cast']+movies['crew']+movies['overview']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id', 'tags', 'title']]
```

```python
new_df
```

```python
# converting all tags to strings
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
# converting all tags to lower case
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return (" ".join(y))
```

```python
# VECTORIZING TEXT
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
# now we have movies in vector form, which is mostly sparse
vectors[0]
```

```python
# stemming to account for similar words
ps.stem('singing')
```

```python
new_df['tags'].apply(stem)
```

```python
len(vectors)
```

```python
len(movies)
```

```python
len(cv.get_feature_names_out())
```

```python
# now we measure the cosine distance between the movie vectors
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[2]
```

```python
sorted(similarity[2], reverse = True)
```

```python
list(enumerate(similarity[2]))
```

```python
# sorting in reverse, top 5 only
sorted(list(enumerate(similarity[0])), reverse = True, key = lambda x:x[1])[1:6]
```

```python
# the recommendation function - sort all movies wrt other movies, decreasing similarities 
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse = True,key = lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_df.iloc[i[0]].title)  #converting index to name
```

```python
recommend('Batman Begins')
```
